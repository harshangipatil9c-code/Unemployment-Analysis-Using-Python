```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/Unemployment in India.csv')

# Preview data
print(df.head())
```

```python
# Rename columns (remove extra spaces if any)
df.columns = df.columns.str.strip()

# Convert date column
df['Date'] = pd.to_datetime(df['Date'])

# Check null values
print(df.isnull().sum())

# Drop missing values (if any)
df = df.dropna()
```

```python
print(df.describe())
print(df['Region'].unique())
```

```python
plt.figure(figsize=(10,5))
df.groupby('Date')['Estimated Unemployment Rate (%)'].mean().plot()
plt.title("Unemployment Rate Over Time (India)")
plt.xlabel("Date")
plt.ylabel("Unemployment Rate (%)")
plt.show()
```

```python
# Create Covid period column
df['Year'] = df['Date'].dt.year

pre_covid = df[df['Year'] == 2019]
covid = df[df['Year'] == 2020]

print("Avg Unemployment (2019):", pre_covid['Estimated Unemployment Rate (%)'].mean())
print("Avg Unemployment (2020):", covid['Estimated Unemployment Rate (%)'].mean())
```

```python
state_data = df.groupby('Region')['Estimated Unemployment Rate (%)'].mean().sort_values()

state_data.plot(kind='barh', figsize=(8,6))
plt.title("State-wise Average Unemployment Rate")
plt.xlabel("Unemployment Rate (%)")
plt.show()
```

```python
df['Month'] = df['Date'].dt.month

monthly_trend = df.groupby('Month')['Estimated Unemployment Rate (%)'].mean()

monthly_trend.plot(marker='o')
plt.title("Monthly Unemployment Trend (Seasonality)")
plt.xlabel("Month")
plt.ylabel("Unemployment Rate (%)")
plt.show()
```

```python
pivot = df.pivot_table(values='Estimated Unemployment Rate (%)',
                       index='Region',
                       columns='Month')

plt.figure(figsize=(10,6))
sns.heatmap(pivot, cmap='coolwarm')
plt.title("Heatmap of Unemployment (Region vs Month)")
plt.show()
```
